```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "mouse_drug_data.csv"
clinical_trial_data_to_load = "clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinical_trial_data_to_load)

#print(mouse_data.head())
#print(clinical_data.head())

# Combine the data into a single dataset

mouse_data_df = pd.DataFrame(mouse_data)
clinical_data_df = pd.DataFrame(clinical_data)

pharma_data = pd.merge(clinical_data_df, mouse_data_df, how = 'outer', on = 'Mouse ID')

# Display the data table for preview

pharma_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

tumor_group = pharma_data.groupby(['Drug','Timepoint'])
#tumor_group.mean()

# Convert to DataFrame

tumor_mean_df = pd.DataFrame(tumor_group.mean())

# Preview DataFrame
#tumor_mean_df.drop(columns = 'Metastatic Sites') #not sure how to remove the metastatic sites column

tumor_mean_df.head() 
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

std_error = pharma_data.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem()

# Convert to DataFrame

std_error_df = pd.DataFrame(std_error)

# Preview DataFrame

std_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
#tumor_munging.drop(columns = 'Metastatic Sites') #supposed to drop the Metastatic Sites column but not sure why it's not
tumor_munging = tumor_mean_df.unstack(0) #unravels dataframe
tumor_munging_df = pd.DataFrame(tumor_munging)
# Preview that Reformatting worked

tumor_munging_df.head()
```

```python
# lists of the tumor volume means per drug
cap_mean_list = tumor_mean_df.loc['Capomulin'].loc[:, 'Tumor Volume (mm3)'].tolist()
inf_mean_list = tumor_mean_df.loc['Infubinol'].loc[:, 'Tumor Volume (mm3)'].tolist()
ket_mean_list = tumor_mean_df.loc['Ketapril'].loc[:, 'Tumor Volume (mm3)'].tolist()
plc_mean_list = tumor_mean_df.loc['Placebo'].loc[:, 'Tumor Volume (mm3)'].tolist()
```

```python
# lists of the tumor volume std error for each drug
cap_std_list = tumor_mean_df.loc['Capomulin'].loc[:, 'Tumor Volume (mm3)'].tolist()
inf_std_list = tumor_mean_df.loc['Infubinol'].loc[:, 'Tumor Volume (mm3)'].tolist()
ket_std_list = tumor_mean_df.loc['Ketapril'].loc[:, 'Tumor Volume (mm3)'].tolist()
plc_std_list = tumor_mean_df.loc['Placebo'].loc[:, 'Tumor Volume (mm3)'].tolist()
```

```python
# Generate the Plot (with Error Bars)

ax = plt.subplot()

x_axis = np.arange(0, 50, 5) # sets axis from 0 to 45 in increments of 5

# Set the plot title and axes titles

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (days)")
plt.ylabel("Tumor Volume (mm3)")

#plots the lines per drug assigning style and color yerr = cap_mean_list, 

ax.errorbar(x_axis, cap_mean_list, yerr = cap_std_list, marker = "o", ls = "dashed", label = "Capomulin",
            color = "red", alpha = 1, capsize = 3)
ax.errorbar(x_axis, inf_mean_list, yerr = inf_std_list, marker = "^", ls = "dashed", label = "Infubinol",
            color = "blue", alpha = 1, capsize = 3)
ax.errorbar(x_axis, ket_mean_list, yerr = ket_std_list, marker = "s", ls = "dashed", label = "Ketapril",
            color = "green", alpha = 1, capsize = 3)
ax.errorbar(x_axis, plc_mean_list, yerr = plc_std_list, marker = "D", ls = "dashed", label = "Placebo",
            color = "black", alpha = 1, capsize = 3)

# Add the legend and gridlines

ax.legend(loc=2)
plt.grid(b = True, which='major', axis='both', linestyle='dotted', linewidth=0.5) 

#not sure where the thick black lines came from or how to get rid of them

plt.ylim(30, 75)
plt.xlim(0, 50)


# Save the Figure

plt.savefig("TumorResponse.png")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

mouse_count = pharma_data.groupby(["Drug","Timepoint"])["Mouse ID"].count()

# Convert to DataFrame

mouse_count_df = pd.DataFrame(mouse_count)

# Preview DataFrame

mouse_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

mouse_munging = mouse_count_df.unstack(0) #unravels dataframe

# Convert to DataFrame

mouse_munging_df = pd.DataFrame(mouse_munging)

# Preview the Data Frame

mouse_munging_df
```

```python
# lists of the survival rates per drug
cap_survival = mouse_count_df.loc['Capomulin'].loc[:, 'Mouse ID'].tolist()
inf_survival = mouse_count_df.loc['Infubinol'].loc[:, 'Mouse ID'].tolist()
ket_survival = mouse_count_df.loc['Ketapril'].loc[:, 'Mouse ID'].tolist()
plc_survival = mouse_count_df.loc['Placebo'].loc[:, 'Mouse ID'].tolist()

#cap_survival
```

```python
#get survival percent
cap_surv_perc = [(x/cap_survival[0])*100 for x in cap_survival]
inf_surv_perc = [(x/inf_survival[0])*100 for x in inf_survival]
ket_surv_perc = [(x/ket_survival[0])*100 for x in ket_survival]
plc_surv_perc = [(x/plc_survival[0])*100 for x in plc_survival]
```

```python
 # Generate the Plot (Accounting for percentages)

ax = plt.subplot()

# Set the x axis from 0 to 45 in increments of 5
x_axis = np.arange(0, 50, 5)

# Set the plot title and axes titles
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")

# Plot the 'mean' list vs. the established x axis with error 
ax.plot(x_axis, cap_surv_perc, marker = "o", ls = "dashed", color = "red", label = "Capomulin")
ax.plot(x_axis, inf_surv_perc, marker = "^", ls = "dashed", color ="blue", label = "Infubinol")
ax.plot(x_axis, ket_surv_perc, marker = "s", ls = "dashed", color = "green", label = "Ketapril")
ax.plot(x_axis, plc_surv_perc, marker = "D", ls = "dashed", color = "black", label = "Placebo")

# Add the legend and gridlines
ax.legend(loc=3)

ax.grid('on', which='major', axis='both', linestyle='dotted', linewidth=0.5)

plt.xlim(0, 50)
        
# Show the resulting scatter plot
plt.savefig("TreatmentSurvival.png")
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug

intial_tumor_volume = 45
percent_change = ((tumor_munging_df.loc[45, :] - intial_tumor_volume) / intial_tumor_volume) * 100
percent_change_drop = percent_change.drop(index = 'Metastatic Sites')
percent_change_drop
```

```python
# Store all Relevant Percent Changes into a Tuple
    
percent_tuple = tuple(zip(percent_change_drop.index, percent_change_drop)) #creates a tuple from percent_change
percent_tuple_list = list(percent_tuple) #converts tuple to a list
#percent_tuple_list

# Splice the data between passing and failing drugs

pass_drugs_list = []
fail_drugs_list = []
index_pass_drugs_list = []
index_fail_drugs_list = []

for x, elements in percent_tuple_list : #loops through list to grab passing and failing drugs
    if elements > 0 :
        pass_drug = elements #stores the pass drugs
        pass_drugs_list.append(elements) #appends the list with the pass drug
        index_pass_drugs_list.append(x) #appends the list with fail drug
    else :
        fail_drugs = elements #stores the fail drugs
        fail_drugs_list.append(elements) #appends the list with the fail drug
        index_fail_drugs_list.append(x) #appends the list with the fail drug
        
pass_drugs = list(zip(index_pass_drugs_list, pass_drugs_list)) #creates a single list of pass drugs
fail_drugs = list(zip(index_fail_drugs_list, fail_drugs_list)) #creates a single list of fail drugs

#fail_drugs[0] #index of Capomulin
#pass_drugs[1, 2, 4]) #indexs of failed drugs

# Orient widths. Add labels, tick marks, etc. 

fig, ax = plt.subplots()
y_pass = [percent_change_drop[2], percent_change_drop[3], percent_change_drop[5]]
y_fail = [percent_change_drop[0]]
x_axis1 = [0]
x_axis2 = [1, 2, 3]

bars1 = ax.bar(x_axis2, y_pass, color='r', alpha=0.8, align="edge", width = -1)
bars2 = ax.bar(x_axis1, y_fail , color='g', alpha=0.8, align="edge", width = -1)

x_labels = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]

plt.setp(ax, xticks=[0, 1, 2, 3], xticklabels=["Capomulin", "Infubinol", "Ketapril", "Placebo"],
        yticks=[-20, 0, 20, 40, 60])
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.grid()

# Call functions to implement the function calls
def autolabel(bars):
    for rect in bars:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2, .1*height, "%d" %int(height)+ "%",
                ha='center', va='top', color="white")

autolabel(bars1)
autolabel(bars2)
fig.tight_layout()

# Save the Figure

plt.savefig("SummaryBar.png")

# Show the Figure

fig.show()
```
