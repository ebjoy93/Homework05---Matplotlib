```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "mouse_drug_data.csv"
clinical_trial_data_to_load = "clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinical_trial_data_to_load)

#print(mouse_data.head())
#print(clinical_data.head())

# Combine the data into a single dataset

mouse_data_df = pd.DataFrame(mouse_data)
clinical_data_df = pd.DataFrame(clinical_data)

pharma_data = pd.merge(clinical_data_df, mouse_data_df, how = 'outer', on = 'Mouse ID')

# Display the data table for preview

pharma_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

tumor_group = pharma_data.groupby(['Drug','Timepoint'])
#tumor_group.mean()

# Convert to DataFrame

tumor_mean_df = pd.DataFrame(tumor_group.mean())

# Preview DataFrame

tumor_mean_df.head() #not sure how to remove the metastatic sites column
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

std_error = pharma_data.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem()

# Convert to DataFrame

std_error_df = pd.DataFrame(std_error)

# Preview DataFrame

std_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
#tumor_munging.drop(columns = 'Metastatic Sites') #supposed to drop the Metastatic Sites column but not sure why it's not
tumor_munging = tumor_mean_df.unstack(0) #unravels dataframe
tumor_munging_df = pd.DataFrame(tumor_munging)
# Preview that Reformatting worked

tumor_munging_df.head()
```

```python
# lists of the tumor volume means per drug
cap_mean_list = tumor_mean_df.loc['Capomulin'].loc[:, 'Tumor Volume (mm3)'].tolist()
inf_mean_list = tumor_mean_df.loc['Infubinol'].loc[:, 'Tumor Volume (mm3)'].tolist()
ket_mean_list = tumor_mean_df.loc['Ketapril'].loc[:, 'Tumor Volume (mm3)'].tolist()
plc_mean_list = tumor_mean_df.loc['Placebo'].loc[:, 'Tumor Volume (mm3)'].tolist()
```

```python
# lists of the tumor volume std error for each drug
cap_std_list = tumor_mean_df.loc['Capomulin'].loc[:, 'Tumor Volume (mm3)'].tolist()
inf_std_list = tumor_mean_df.loc['Infubinol'].loc[:, 'Tumor Volume (mm3)'].tolist()
ket_std_list = tumor_mean_df.loc['Ketapril'].loc[:, 'Tumor Volume (mm3)'].tolist()
plc_std_list = tumor_mean_df.loc['Placebo'].loc[:, 'Tumor Volume (mm3)'].tolist()
```

```python
# Generate the Plot (with Error Bars)

ax = plt.subplot()

x_axis = np.arange(0, 50, 5) # sets axis from 0 to 45 in increments of 5

# Set the plot title and axes titles

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (days)")
plt.ylabel("Tumor Volume (mm3)")

#plots the lines per drug assigning style and color yerr = cap_mean_list, 

ax.errorbar(x_axis, cap_mean_list, yerr = cap_std_list, fmt = "o", ls = "dashed", label = "Capomulin",
            color = "red", alpha = 1, capsize = 3)
ax.errorbar(x_axis, inf_mean_list, yerr = inf_std_list, fmt = "^", ls = "dashed", label = "Infubinol",
            color = "blue", alpha = 1, capsize = 3)
ax.errorbar(x_axis, ket_mean_list, yerr = ket_std_list, fmt = "s", ls = "dashed", label = "Ketapril",
            color = "green", alpha = 1, capsize = 3)
ax.errorbar(x_axis, plc_mean_list, yerr = plc_std_list, fmt = "D", ls = "dashed", label = "Placebo",
            color = "black", alpha = 1, capsize = 3)

# Add the legend and gridlines

ax.legend(loc=2)
plt.grid(b = True, which='major', axis='both', linestyle='dotted', linewidth=0.5) 

#not sure where the thick black lines came from or how to get rid of them

plt.ylim(30, 75)
plt.xlim(0, 50)


# Save the Figure

plt.savefig("TumorResponse.png")
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 

metastatic_group = pharma_data.groupby(["Drug","Timepoint"])
#metastatic_group["Metastatic Sites"].mean()

# Convert to DataFrame

metastatic_mean_df = pd.DataFrame(metastatic_group["Metastatic Sites"].mean())

# Preview DataFrame

metastatic_mean_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

std_error_meta = pharma_data.groupby(["Drug","Timepoint"])["Metastatic Sites"].sem()

# Convert to DataFrame

std_error_meta_df = pd.DataFrame(std_error_meta)

# Preview DataFrame

std_error_meta_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

meta_munging = metastatic_mean_df.unstack(0) #unravels dataframe
meta_munging_df = pd.DataFrame(meta_munging)

# Preview that Reformatting worked

meta_munging_df.head()
```

```python
# lists of the metastatic sites means per drug
cap_mean_meta = metastatic_mean_df.loc['Capomulin'].loc[:, 'Metastatic Sites'].tolist()
inf_mean_meta = metastatic_mean_df.loc['Infubinol'].loc[:, 'Metastatic Sites'].tolist()
ket_mean_meta = metastatic_mean_df.loc['Ketapril'].loc[:, 'Metastatic Sites'].tolist()
plc_mean_meta = metastatic_mean_df.loc['Placebo'].loc[:, 'Metastatic Sites'].tolist()
```

```python
# lists of the metastatic sites std error for each drug
cap_std_meta = metastatic_mean_df.loc['Capomulin'].loc[:, 'Metastatic Sites'].tolist()
inf_std_meta = metastatic_mean_df.loc['Infubinol'].loc[:, 'Metastatic Sites'].tolist()
ket_std_meta = metastatic_mean_df.loc['Ketapril'].loc[:, 'Metastatic Sites'].tolist()
plc_std_meta = metastatic_mean_df.loc['Placebo'].loc[:, 'Metastatic Sites'].tolist()
```

```python
# Generate the Plot (with Error Bars)

ax = plt.subplot()

x_axis = np.arange(0, 50, 5) # sets axis from 0 to 45 in increments of 5

# Set the plot title and axes titles

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (days)")
plt.ylabel("Met. Sites")

#plots the lines per drug assigning style and color yerr = cap_mean_list, 

ax.errorbar(x_axis, cap_mean_meta, yerr = cap_std_meta, fmt = "o", ls = "dashed", label = "Capomulin",
            color = "red", alpha = 1, capsize = 5)
ax.errorbar(x_axis, inf_mean_meta, yerr = inf_std_meta, fmt = "^", ls = "dashed", label = "Infubinol",
            color = "blue", alpha = 1, capsize = 5)
ax.errorbar(x_axis, ket_mean_meta, yerr = ket_std_meta, fmt = "s", ls = "dashed", label = "Ketapril",
            color = "green", alpha = 1, capsize = 5)
ax.errorbar(x_axis, plc_mean_meta, yerr = plc_std_meta, fmt = "D", ls = "dashed", label = "Placebo",
            color = "black", alpha = 1, capsize = 5)

# Add the legend and gridlines

ax.legend(loc=2)
plt.grid(b = True, which='major', axis='both', linestyle='dotted', linewidth=0.5)  

#not sure where the thick black lines came from or how to get rid of them

plt.ylim(0.0, 4.0)
plt.xlim(0,50)


# Save the Figure

plt.savefig("MetaSitesResponse.png")
```
